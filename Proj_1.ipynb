```python
import pandas as pd

df = pd.read_csv(r'C:\Users\User\Downloads\Bengaluru_House_Data.csv')
df
```

```python
df = df.drop(['society','availability'],axis='columns')
df
```

```python
df['balcony'] = df['balcony'].fillna(0.0)
df
```

```python
df = df.dropna()
df
```

```python
df.loc[:,'bhk'] = df['size'].apply(lambda x : x.split(' ')[0]).astype(float)
df1 = df.drop('size',axis='columns')
df1
```

```python
def flo(x):
    try:
        float(x)
    except:
        return False
    return True

df1[~df1['total_sqft'].apply(flo)]
```

```python
def mean(x):
    values = x.split("-")
    if len(values) == 1:
        try:
            return float(x)
        except:
            return None
    elif len(values) == 2:
        return (float(values[0])+float(values[1]))/2
    else:
        return None
        

df1['total_sqft'] = df1['total_sqft'].apply(mean)
df1
```

```python
df1 = df1.dropna()
df1
```

```python
df1.isnull().sum()
```

```python
len(df1['location'].unique())
```

```python
df1.loc[:,'location'] = df1['location'].apply(lambda x: x.strip())
loc_data = df1.groupby('location')['location'].agg('count').sort_values(ascending=False)
loc_data
```

```python
from matplotlib import pyplot as map

lo_leng = []
data_points = []
for i in range(1,30):
    count = sum(loc_data[loc_data<=i])
    lo_leng.append(count)
    data_points.append(i)

map.scatter(data_points,lo_leng)
map.xlabel('data points')
map.ylabel('number of properties')
map.show
```

```python
waste_loc = loc_data[loc_data<=10]
print(sum(waste_loc))
# if 'Richards Town' in waste_loc:
#     print(loc_data['Richards Town'])
```

```python
df1.loc[:,'location'] = df1['location'].apply(lambda x: 'other' if x in waste_loc else x)
df1
```

```python
len(df1['location'].unique())
```

```python
df2 = df1.rename(columns={'price':'price_(lacs)'})
df2['price/sqft'] = df1['price']*100000/df1['total_sqft']
df2
```

```python
# df2['total_sqft'].describe()



# flag = 0
# ar = []
# for area in df2['total_sqft']:
#     if area <= 10:
#         flag = flag + 1
#         ar.append(area)
    
# print(flag, ar)



df3 = df2.drop(df2[df2['total_sqft']<=10].index)
df3
```

```python
# df3['total_sqft'].describe()

LB = df3['total_sqft'].mean() - 2*df3['total_sqft'].std()
UB = df3['total_sqft'].mean() + 2*df3['total_sqft'].std()
df4 = df3.drop(df3[df3['total_sqft']<LB].index)
df4 = df3.drop(df3[df3['total_sqft']>UB].index)
df4
```

```python
# df4['total_sqft'].describe()
df5 = df4.drop(df4[df4['balcony']>(df4['bhk']+2)].index)
df6 = df5.drop(df5[df5['bath']>(df5['bhk']+2)].index)
df6
```

```python
df7 = df6.drop(df6[(df6['total_sqft']/df6['bhk'])<300].index)
df7
```

```python
df7['price/sqft'].describe()

ub = df7['price/sqft'].mean() + 3*df7['price/sqft'].std()
lb = df7['price/sqft'].mean() - 3*df7['price/sqft'].std()

df8 = df7.drop(df7[df7['price/sqft']<=lb].index)
df8 = df7.drop(df7[df7['price/sqft']>=ub].index)

df8
```

```python
for loc,ldata in df8.groupby('location'):
    bhk_det = {}
    for bhk,bhk_data in ldata.groupby('bhk'):
        bhk_det[bhk] = {'mean' : bhk_data['price_(lacs)'].mean(),
                         'sd' : bhk_data['price_(lacs)'].std(),
                         'count' : bhk_data['price_(lacs)'].count()}

    for bhk,bhk_data in ldata.groupby('bhk'):
        mean_price = bhk_det.get(bhk-1,{}).get('mean',0)

        df8 = df8.drop(bhk_data[bhk_data['price_(lacs)'] <= mean_price].index)
df8
```

```python
df9 = df8.drop('price/sqft',axis='columns')
dummies1 = pd.get_dummies(df8['area_type']).astype(int)
dummies2 = pd.get_dummies(df8['location']).astype(int)

df10 = pd.concat([df9,dummies1.drop('Super built-up  Area',axis='columns'),dummies2.drop('other',axis='columns')],axis='columns')
df11 = df10.drop(['area_type','location'],axis='columns')
df11
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
# from sklearn.tree import DecisionTreeRegressor
# from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import ShuffleSplit

x = (df11.drop('price_(lacs)',axis='columns'))
y = (df11['price_(lacs)'])

algos = {
    'linear regression' : {
        'model' : LinearRegression(),
        'params' : {
            'fit_intercept' : [True, False],
            'copy_X' : [True, False]

        }
    },

    'Lasso' : {
        'model' : Lasso(),
        'params' : {
            'selection' : ['cyclic','random'],
            'alpha' : [1,2]
        }
    }

    # 'DecisionTree' : {
    #     'model' : DecisionTreeRegressor(),
    #     'params' : {
    #         'criterion' : ['friedman_mse','absolute_error'],
    #         'max_depth': [None, 10, 20, 30]
    #     }
    # },

    # 'SVR' : {
    #     'model' : SVR(),
    #     'params' : {
    #         'kernel' : ['linear','rbf'],
    #         'C' : [1,5,10,15,20],
    #         'gamma' : ['auto','scale']
    #     }
    # }
}

DF = {}

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

for model_name, info in algos.items():
    gs = GridSearchCV(info['model'],info['params'],cv=cv, error_score='raise')
    gs.fit(x,y)
    DF[model_name] = {
        'best score' : gs.best_score_,
        'best param' : gs.best_params_
    }
pd.DataFrame(DF)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)

LR = LinearRegression(fit_intercept=True, copy_X=True)

LR.fit(x_train,y_train)
LR.score(x_test,y_test)
```

```python
import numpy as np

location_columns = x.columns[7:]
print(pd.DataFrame(location_columns, columns=['location']))

Area_columns = x.columns[4:7]
print(pd.DataFrame(Area_columns, columns=['area']))

def propertyPredict(x, location_index, area_index, totSqft, bath, balcony, bhkk):
    # Initialize feature array with zeros
    features = np.zeros(len(x.columns))
    
    if 0 <= location_index < len(location_columns):
        features[7 + location_index] = 1
    
    if 0 <= area_index < len(Area_columns):
        features[4 + area_index] = 1

    ## for super built up area and others in location
    # we enter a number greater than their limits so that it never enters the loop

    
    # Set other features
    features[0] = totSqft
    features[1] = bath
    features[2] = balcony
    features[3] = bhkk
    
    # Create a DataFrame for prediction
    X = pd.DataFrame([features], columns=x.columns)
    
    # Predict using the model
    prediction = LR.predict(X)
    
    return prediction
```
